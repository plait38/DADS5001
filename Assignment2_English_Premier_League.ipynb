## simple stats 3 questions

1.Mean Half Time Away Team Goals of each Team

```python
df2 = pd.DataFrame({"mean":df1.groupby("AwayTeam").HTAG.mean()})
df2
```

2. Top 5 HomeTeams got red card



```python
HTeam = pd.DataFrame(df1.groupby("HomeTeam").HR.sum())
ATeam = pd.DataFrame(df1.groupby("AwayTeam").AR.sum())
rs= pd.merge(HTeam, ATeam, left_index=True, right_index=True)
rs1 = pd.DataFrame ({'Total_redcard':rs['HR']+rs['AR']})
rs1.sort_values(by="Total_redcard",ascending=False).head(5)
```

3.Referee who is biased to Home and Away teams



```python
rf = df1.groupby('Referee')[["HR","AR","HY","AY"]].sum()
rf_sum = pd.DataFrame({"HR": rf['HR'],
                       "AR": rf['AR'],
                       "HY": rf['HY'],
                       "AY": rf['AY'],
                       "Total_card":rf['HR']+rf['AR']+rf['HY']+rf['AY'],
                       })
rf_sum.sort_values(by='Total_card',ascending=False)
```

## **advanced 2 question**

Which HomeTeam have most Shots Off Target percentage?

```python
HT = df1.groupby("HomeTeam")[['HS','HST']].sum()
#HF = (HT['HS']-HT['HST'])
HF = ((HT['HS']-HT['HST'])*100/(HT['HS'])).reset_index(name = "HF%")
HFR = pd.merge(HT,HF, on ="HomeTeam")
HFR.sort_values(by ="HF%",ascending=False)
```

Which team and match should we bet on to win?
- check got most Bet365 home win odds percentage

```python
WB = pd.DataFrame(df1.loc[:,["HomeTeam","AwayTeam","B365H","B365D","B365A"]])
BP = pd.DataFrame({"HomeTeam":WB.loc[:,"HomeTeam"],
                   "AwayTeam":WB.loc[:,"AwayTeam"],
                   "B365H" : WB.loc[:,"B365H"],
                   "B365D" : WB.loc[:,"B365D"],
                   "B365A" : WB.loc[:,"B365A"],
                   "BetsHome_win%" : WB['B365H']*100/(WB['B365H']+WB['B365D']+WB['B365A'])})
BP.sort_values(by='BetsHome_win%',ascending = False)
```
