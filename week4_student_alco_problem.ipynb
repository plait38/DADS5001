### Step 1. Import the necessary libraries

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/Students_Alcohol_Consumption/student-mat.csv).

```python
data = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/Students_Alcohol_Consumption/student-mat.csv")
data
```

### Step 3. Assign it to a variable called df.

```python
df = pd.DataFrame(data)
df
```

### Step 4. For the purpose of this exercise slice the dataframe from 'school' until the 'guardian' column

```python
df.iloc[:, :12]
```

### Step 5. Create a lambda function that will capitalize strings.

```python
df1 = df.applymap(lambda x: str(x).capitalize())
df1
```

### Step 6. Use the lambda function created from step 5 to capitalize both Mjob and Fjob

```python
df['Mjob'] = df['Mjob'].apply(lambda x: str(x).capitalize())
df['Fjob'] = df['Fjob'].apply(lambda x: str(x).capitalize())
df
```

### Step 7. Create a lambda function called majority that returns a boolean value to a new column called legal_drinker (Consider majority as older than 17 years old)

```python
df['legal_drinker'] = df['age'].apply(lambda x: True if x >= 17 else False)
df
```
